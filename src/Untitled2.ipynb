```python
import pandas as pd
mnist_csv = pd.read_csv( './mnist_train.csv', header=None, skiprows =1).values
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split( mnist_csv, test_size=0.3, random_state=1 )
```
